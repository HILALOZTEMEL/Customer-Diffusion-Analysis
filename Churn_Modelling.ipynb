```python

#1.kutuphaneler
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#2.veri onisleme
#2.1.veri yukleme

#kimlik doğrulama işlemi :drive in içinde çalışma ortamı sağlarız serverlarda
from google.colab import drive
drive.mount('/gdrive')
%cd /gdrive

# Load Dataset
df = pd.read_csv('/gdrive/My Drive/hilalCalisma/Müsteri yayılım analizi/Churn_Modelling.csv')



#test
df.head()
```

```python

#veri on isleme
X= df.iloc[:,3:13].values
y = df.iloc[:,13].values


```

```python
#encoder: Kategorik -> Numeric
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
X[:,1] = le.fit_transform(X[:,1])#Geography

le2 = preprocessing.LabelEncoder()
X[:,2] = le2.fit_transform(X[:,2])#Gender

from sklearn.compose import ColumnTransformer #Birden fazla kolonun aynı anda dönüşümünü sağlar, dönüşen öz nitelik tek bir öz nitelik alanında birleşmsini sağlar
from sklearn.preprocessing import OneHotEncoder

ohe = ColumnTransformer([("ohe", OneHotEncoder(dtype=float),[1])],
                        remainder="passthrough"
                        )
X = ohe.fit_transform(X)
X = X[:,1:]
```

```python
#verilerin egitim ve test icin bolunmesi
from sklearn.model_selection import train_test_split

x_train, x_test,y_train,y_test = train_test_split(X,y,test_size=0.33, random_state=0)

#verilerin olceklenmesi (0 ile 1 aralığında)
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

X_train = sc.fit_transform(x_train)
X_test = sc.fit_transform(x_test)
```

```python
#3 Yapay Sinir ağı
import keras
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential()

classifier.add(Dense(6, kernel_initializer = 'uniform', activation = 'relu' , input_dim = 11))

classifier.add(Dense(6, kernel_initializer = 'uniform', activation = 'relu'))

classifier.add(Dense(1, kernel_initializer = 'uniform', activation = 'sigmoid'))

classifier.compile(optimizer = 'adam', loss =  'binary_crossentropy' , metrics = ['accuracy'] )

classifier.fit(X_train, y_train, epochs=50)

y_pred = classifier.predict(X_test)

y_pred = (y_pred > 0.5)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)

print(cm)
```
